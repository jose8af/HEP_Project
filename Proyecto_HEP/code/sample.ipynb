```python
import matplotlib.pyplot as plt
import uproot
import numpy as np
```

```python
nano_dy = "/Users/jochoafl/cms10_sim/CMSSW_10_6_30/src/NanoAOD.root"
```

```python
nano_dy_root = uproot.open(f"{nano_dy}:Events")
```

```python
muon_pt = nano_dy_root["Muon_pt"].array()
muon_phi = nano_dy_root["Muon_phi"].array()
muon_eta = nano_dy_root["Muon_eta"].array()
numbermuon = nano_dy_root["nMuon"].array()

numbermuon = np.array(numbermuon)
index_dimuon = np.where(numbermuon >= 2)[0]    

ele_pt = nano_dy_root["Electron_pt"].array()
ele_phi = nano_dy_root["Electron_phi"].array()
ele_eta = nano_dy_root["Electron_eta"].array()
numberele = nano_dy_root["nElectron"].array()

numbermuon = np.array(numbermuon)
index_dimuon = np.where(numbermuon >= 2)[0]   

numberele = np.array(numberele)
index_diele = np.where(numberele >= 2)[0]   
```

```python
dm_imass=[]   
for i in index_dimuon:
    imass=(2*muon_pt[i][0]*muon_pt[i][1]*(np.cosh(muon_eta[i][0]-muon_eta[i][1])-np.cos(muon_phi[i][0]-muon_phi[i][1])))**(1/2)
    dm_imass.append(imass)

dm_imass_ele=[]
for i in index_diele:
    imass=(2*ele_pt[i][0]*ele_pt[i][1]*(np.cosh(ele_eta[i][0]-ele_eta[i][1])-np.cos(ele_phi[i][0]-ele_phi[i][1])))**(1/2)
    dm_imass_ele.append(imass)
```

```python
plt.hist(dm_imass, bins='auto', histtype='step', fill=False)  
plt.title("Invariant mass ")
plt.xlim([0,300])
plt.xlabel('Dimuon mass [GeV]')
plt.ylabel('Number of Events')
plt.show()
```

```python
plt.hist(dm_imass_ele, bins='auto', histtype='step', fill=False)  
plt.title("Invariant mass ")
plt.xlim([0,300])
plt.xlabel('Dimuon mass [GeV]')
plt.ylabel('Number of Events')
plt.show()
```
